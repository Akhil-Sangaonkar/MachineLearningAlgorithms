# Perceptron — Binary Classification

## 1. Problem Setup

Assume we have a dataset

$$
\{(x_i,y_i)\}_{i=1}^{N}
$$

where

$$
x_i \in \mathbb{R}^D
$$

is the feature vector and

$$
y_i \in \{-1, +1\}
$$

is the binary class label.

The goal is to classify each input into one of two classes.

---

## 2. Linear Model

The perceptron learns a single weight vector

$$
w \in \mathbb{R}^D
$$

The score for an input is

$$
s = w^T x
$$

---

## 3. Adding the Bias Term

To include an intercept, we augment the feature vector

$$
x =
\begin{bmatrix}
1 \\
x
\end{bmatrix}
$$

The weight vector becomes

$$
w =
\begin{bmatrix}
b \\
w
\end{bmatrix}
$$

Thus the score becomes

$$
s = w^T x
$$

---

## 4. Prediction Rule

The predicted class is determined by the sign of the score

$$
\hat{y} =
\begin{cases}
+1 & w^T x \ge 0 \\
-1 & w^T x < 0
\end{cases}
$$

---

## 5. Objective Function

The perceptron does not minimize a smooth loss.

Instead, it updates weights only when a sample is misclassified:

$$
\text{if } y_i (w^T x_i) \le 0
$$

then update the weights.

---

## 6. Update Rule

For a misclassified sample $(x_i, y_i)$:

$$
w = w + \eta y_i x_i
$$

where

$$
\eta
$$

is the learning rate.

---

## 7. Batch Update 

For multiple misclassified samples, the update becomes

$$
w =
w + \eta \sum_{i \in \mathcal{M}} y_i x_i
$$

where

$$
\mathcal{M} = \{i : y_i \neq \hat{y}_i\}
$$

is the set of misclassified indices.

---

## 8. Initialization

The weights are initialized as

$$
w = 0
$$

At this point all predictions are

$$
\hat{y} = +1
$$

---

## 9. Iterative Optimization

For each epoch:

Compute predictions

$$
\hat{y} = \text{sign}(Xw)
$$

Find misclassified samples

$$
\mathcal{M} = \{i : \hat{y}_i \neq y_i\}
$$

Update weights

$$
w = w + \eta \sum_{i \in \mathcal{M}} y_i x_i
$$

---

## 10. Convergence Criterion

The algorithm stops when

$$
\hat{y}_i = y_i \quad \forall i
$$

i.e., no misclassifications remain.

This happens when the data is **linearly separable**.

---

## 11. Geometric Interpretation

The perceptron learns a **linear decision boundary**

$$
w^T x = 0
$$

Points satisfying

$$
w^T x > 0
$$

are classified as $+1$, and

$$
w^T x < 0
$$

as $-1$.

---

## 12. Effect of Learning Rate

The update rule is

$$
w = w + \eta y x
$$

Scaling $\eta$ results in

$$
\eta \uparrow \Rightarrow \|w\| \uparrow
$$

but the **decision boundary remains unchanged** since it depends only on direction.

---

## 13. Prediction

After training, predictions are computed as

$$
\hat{y} = \text{sign}(w^T x)
$$

---

## 14. Algorithm Summary

Initialize

$$
w = 0
$$

Repeat for each epoch:

Compute predictions

$$
\hat{y} = \text{sign}(Xw)
$$

Find misclassified samples

$$
\mathcal{M} = \{i : \hat{y}_i \neq y_i\}
$$

Update weights

$$
w = w + \eta \sum_{i \in \mathcal{M}} y_i x_i
$$

Stop if no misclassifications remain.

---

## 15. Final Optimization Perspective

The perceptron implicitly solves

$$
\text{find } w \text{ such that } y_i (w^T x_i) > 0
$$

for all training samples.

This produces a classifier that separates two classes using a **linear decision boundary**.

```python
class Perceptron:
    def __init__(self, learning_rate =0.01, epochs=100, add_bias =True):
        self.learning_rate = learning_rate
        self.epochs = epochs
        self.add_bias = add_bias

        self.weights = None


    def fit(self,X,y):
        X = np.asarray(X)
        y = np.asarray(y)

        y = y.reshape(-1)

        if X.ndim==1:
            X = X.reshape(-1,1)

        N , D = X.shape

        if self.add_bias :
            X = np.column_stack((np.ones(N),X))
            D +=1

        self.weights = np.zeros(D)


        for epoch in range(self.epochs):

            y_pred = np.where((X@ self.weights ) >=0,1,-1)

            mistakes = np.where(y_pred != y)[0]
            

            if len(mistakes) >0 :
                self.weights += self.learning_rate * np.sum(y[mistakes, None] * X[mistakes,:],axis=0)

            

            if len(mistakes) ==0:
                print(f'Epoch {epoch}: converged (0 mistakes)')  
                print(f'The model weights are : \n {self.weights}')
                break
                
     


    def predict(self,X):
        X = np.asarray(X)
    
    
        if X.ndim==1:
            X = X.reshape(-1,1)

        N = len(X)
    
    
        if self.add_bias :
            X = np.column_stack((np.ones(N),X))


        y_pred = np.where((X @ self.weights ) >=0,1,-1)

        return y_pred
          
```



## 1. Linearly Separable Dataset

We construct a dataset such that

$$
y = \text{sign}(w^T x)
$$

where the true weights are known.

---

## 2. Effect of Learning Rate Scaling

The perceptron update rule is

$$
w = w + \eta y x
$$

Scaling $\eta$ scales the weights but does not change predictions.

---



```python
import numpy as np

np.random.seed(42)

N = 100

X_class1 = np.random.randn(N,2) + np.array([3,3])
X_class2 = np.random.randn(N,2) + np.array([-3,3])
X_train = np.vstack([X_class1, X_class2])
y_train = np.array([1]*N + [-1]*N )

```

---

## 3. Expected Behaviour

- Model converges for linearly separable data
- Larger $\alpha$ leads to larger weights
- Predictions remain identical
---

```python
print("learning_rate = 0.01")

model_1 = Perceptron(learning_rate=0.01, epochs=1000, add_bias=False)
model_1.fit(X_train,y_train)

pred_1 = model_1.predict(X_train)

print("\nTraining Accuracy:", np.mean(pred_1 == y_train))
```

```python
print("learning_rate = 0.1")

model_2 = Perceptron(learning_rate=0.1, epochs=1000, add_bias=False)
model_2.fit(X_train,y_train)

pred_2 = model_2.predict(X_train)

print("\nTraining Accuracy:", np.mean(pred_2 == y_train))
```

```python
print("Weight Scaling Comparison")

norm_1 = np.linalg.norm(model_1.weights)
norm_2 = np.linalg.norm(model_2.weights)

print("||W (0.01)||:", norm_1)
print("||W (0.1)|| :", norm_2)

print("\nRatio (~10 expected):", norm_2 / norm_1)
```

```python
print("Prediction Consistency")

print("Predictions identical?", np.all(pred_1 == pred_2))
```
